```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.constants as con
```

```python
Dt=np.loadtxt('IRCF.txt')
h=con.Planck
k=con.Boltzmann
da=con.c
x=Dt[:,0:1]  
y=Dt[:,1:2]  
z=Dt[:,3:4]
plt.plot(x,y,'-o')
```

```python
for i in range (42):
    y[i] = y[i]/.0001
    z[i] = z[i]*da*100
plt.errorbar(x,y,z)
```

```python
def I(v,T):
    p1=h*(pow(v,3))
    p2 = pow(da,2)
    p3 = (h*v)/(k*T)
    return (p1/(p2*(np.exp(p3)-1)))
for i in range (42):
    u=I(x[i],250)
```
