```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.constants as con
```

```python
Dt=np.loadtxt('IRCF.txt')
h=con.Planck
k=con.Boltzmann
da=con.c
x=Dt[:,0:1]  
y=Dt[:,1:2]  
z=Dt[:,3:4]
plt.plot(x,y,'-o')
```

```python
for i in range (42):
    z[i] = z[i]/.0001 ##Se transforman las unidades de kilo jansky a mega jansky
    x[i] = x[i]*da*1000 ##Se hace la convercion de la frecuencia con la constante la velocidad e la luz
plt.errorbar(x,y,z) ##Aqui se graficó los errores dados en la hoja de datos
```

```python
def I(v,T):
    p1=h*(pow(v,3))##Se divide en tres partes la formula para tener la mejor manipulación si existe algún error
    p2 = pow(da,2)
    p3 = (h*v)/(k*T)
    return (p1/(p2*(np.exp(p3)-1)))
u = np.zeros(42)
f =np.linspace(1,5,100)##Son las diferentes temperatras que se utilizaran pra evaluar
for i in range (42):
    u=I(x,f[i])##Se hará una valoración con todas las diferente frecuencias
    plt.plot(x,u)
```
