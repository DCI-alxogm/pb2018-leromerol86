```python
import numpy as np
import matplotlib.pyplot as plta
import math as math
```

$\mathbf {Se \ asignan \ valores \ a \ las variables, \ y  \ se \ inicializan \ los \ arreglos. }$

```python
#Se asigan un valor entre cada tiempo de 0.01 para asi realizar diferentes cantidades puntos
h=.01
t=np.arange(1,10,h)#se los puntos del tiempo
v=np.zeros(len(t))#se inicializan las variables en 0, en toda su dimensión
theta=np.zeros(len(t))
v[0]=10 #Valores iniciales de la velocidad
theta[0]=500
w=5#velocidad angular
```

```python
for i in range (800):
    theta[i+1]=theta[i]+(h*v[i])
    v[i+1]=v[i]+(h*(-w**2*theta[i]))
plt.plot(t,v)#IMpresión de la grafica, con los valores del tiempo ,contra la velocidad
```

$\mathbf{Gráfica\ de \ tiempo \ contra \ theta}$

```python
plt.plot(t,theta)
```

$\mathbf{Gráfica\ de \ velocidad \ contra \ theta}$

```python
plt.plot(v,theta)
```
